### Un example concret  : Fibonnaci

La suite de Fibonacci est définie par la relation de récurrence suivante :

$$ F(n) = F(n-1) + F(n-2) $$
avec les conditions initiales $ F(0) = 0 $ et $ F(1) = 1 $.

Et donc : 

$$ F(2) = F(1) + F(0) = 1 + 0 = 1 $$
$$ F(3) = F(2) + F(1) = 1 + 1 = 2 $$
$$ F(4) = F(3) + F(2) = 2 + 1 = 3 $$

````python
def fib(n) :
    if n == 1 : return 1 
    if n == 0 : return 0
    return fib(n-1) + fib(n-2)

fib(4)
````

````python
# Une librairie intéressante functools 

from functools import lru_cache


@lru_cache(maxsize=None)
def fib(n) :
    if n == 1 : return 1 
    if n == 0 : return 0
    return fib(n-1) + fib(n-2)

fib(1000)
````

## Args & Kwargs



````python
def my_func(a, b, c=1, d=2) :
    print(a, b, c, d)

my_func("A", b="B")
# SI je met un => c'est un argument nommé, sinon c'est un argument positionnel A partir de la je doit mettre que des =
my_func("A", "B",1, 2)
````

````python
def my_func(a, b, *args) :
    print(type(args))
    print(a, b, args)
    if args :
        print("Args : ", args[0])
    
my_func("A", "B")
my_func("A", "B", 1, 2, 3, "bob","renard")
````

````python
def my_func(c= 1, d = 2 , **kwargs) :
    print(type(kwargs))
    print(c, d, kwargs)
    if kwargs :
        print("Args : ", kwargs["name"])
    
my_func(c = "A", d = "B")
my_func(c = "A", d = "B", name="Bob", age=30)
````

````python
def func(a, b, *args, c=1, d=2, **kwargs) :
    
    print("Argument", a, b) 
    print("Args :", args)
    print("Keyword Args :", kwargs)
    
func(1, 2, 3, 4, 5, c=10, d=20, name="Bob", age=30)
````

````python
# Décorateur amélioré
def time_it(func) :
    def wrapper(*args, **kwargs):
        start = time.time()
        # time.sleep(1)
        func(*args, **kwargs)
        end = time.time()
        print(f"Final Time {(end - start):.2f} s")
    return wrapper

@time_it
def compute(n):
    j = 0
    for i in big_cat_bar(range(n)): 
        j += 1
compute(10_000)
````
